## 6. Cluster the publish time into 10-minute intervals (e.g. from 02:20 to 02:30)


Convert into datetime

```python
tot['publish_time'] = pd.to_datetime(tot['publish_time'])
```

We extract only hours and minutes

```python
tot['time_only'] = tot['publish_time'].dt.time
```

We create a column for 10-minute intervals and then we go to create the intervals,
extracting the minutes from midnight (from 0 to 1439) we divide by 10 minutes creating the intervals

```python
tot['minutes_from_midnight'] = tot['publish_time'].dt.hour * 60 + tot['publish_time'].dt.minute
tot['interval_10min'] = (tot['minutes_from_midnight'] // 10) * 10

def format_interval(minutes):
    start_hour = minutes // 60
    start_min = minutes % 60
    end_minutes = minutes + 10
    end_hour = (end_minutes // 60) % 24
    end_min = end_minutes % 60
    return f"{start_hour:02d}:{start_min:02d}-{end_hour:02d}:{end_min:02d}"
tot['interval_label'] = tot['interval_10min'].apply(format_interval)
```

## 7. For each interval, determine the number of videos, average number of likes and of dislikes.

We calculate for each interval the number of videos, the average of likes and dislikes, 
then we add the labels and reorder the columns.

```python
# De-duplication: for each video we keep only the last one
tot_7 = tot.sort_values(['video_id', 'trending_date'])
tot_7 = tot_7.drop_duplicates(subset=['video_id', 'publish_time', 'channel_title'], keep='last')

# Aggregation
interval_stats = tot_7.groupby('interval_10min').agg(
    num_videos=('video_id', 'count'),
    avg_likes=('likes', 'mean'),
    avg_dislikes=('dislikes', 'mean')
)

interval_stats['interval'] = interval_stats.index.map(format_interval)
interval_stats = interval_stats[['interval', 'num_videos', 'avg_likes', 'avg_dislikes']]
interval_stats
```
